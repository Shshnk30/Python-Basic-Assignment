```python
#Ans-4 #This overview covers the main types of operators in Python with examples demonstrating their use.

# Arithmetic Operators
a = 5
b = 3

# Addition
print(a + b)  # Output: 8

# Subtraction
print(a - b)  # Output: 2

# Multiplication
print(a * b)  # Output: 15

# Division (returns float)
print(a / b)  # Output: 1.666...

# Modulus (remainder of division)
print(a % b)  # Output: 2

# Exponentiation (power)
print(a ** b)  # Output: 125

# Floor Division (integer division)
print(a // b)  # Output: 1

# Comparison Operators
# Equal to
print(a == b)  # Output: False

# Not equal to
print(a != b)  # Output: True

# Greater than
print(a > b)  # Output: True

# Less than
print(a < b)  # Output: False

# Greater than or equal to
print(a >= b)  # Output: True

# Less than or equal to
print(a <= b)  # Output: False

# Logical Operators
# AND (both conditions must be True)
print(a > 0 and b > 0)  # Output: True

# OR (at least one condition must be True)
print(a > 0 or b < 0)  # Output: True

# NOT (inverts the condition)
print(not (a > 0))  # Output: False

# Assignment Operators
# Assignment
c = a + b

# Addition Assignment
c += 2  # Equivalent to c = c + 2
print(c)  # Output: 10

# Subtraction Assignment
c -= 1  # Equivalent to c = c - 1
print(c)  # Output: 9

# Multiplication Assignment
c *= 2  # Equivalent to c = c * 2
print(c)  # Output: 18

# Division Assignment
c /= 2  # Equivalent to c = c / 2
print(c)  # Output: 9.0

# Bitwise Operators
# AND (bitwise)
print(5 & 3)  # Output: 1

# OR (bitwise)
print(5 | 3)  # Output: 7

# XOR (bitwise)
print(5 ^ 3)  # Output: 6

# Complement (bitwise NOT)
print(~5)  # Output: -6

# Left Shift (shifts bits to the left)
print(5 << 1)  # Output: 10

# Right Shift (shifts bits to the right)
print(5 >> 1)  # Output: 2

# Membership Operators
# In (checks if a value is in a sequence)
print(3 in [1, 2, 3])  # Output: True

# Not In (checks if a value is not in a sequence)
print(4 not in [1, 2, 3])  # Output: True

# Identity Operators
# Is (checks if two variables refer to the same object)
a = [1, 2, 3]
b = a
print(a is b)  # Output: True

# Is Not (checks if two variables refer to different objects)
c = [1, 2, 3]
print(a is not c)  # Output: True
```

```python
#Ans-5 #the concept of type casting in Python examples are-

# Arithmetic Type Casting Functions

# Convert a float to an integer (removes the decimal part)
num = 3.14
int_num = int(num)
print(int_num)  # Output: 3

# Convert a string to an integer
str_num = "42"
int_str_num = int(str_num)
print(int_str_num)  # Output: 42

# Convert an integer to a float
num = 5
float_num = float(num)
print(float_num)  # Output: 5.0

# Convert a string to a float
str_num = "3.14"
float_str_num = float(str_num)
print(float_str_num)  # Output: 3.14

# Convert an integer to a string
num = 10
str_num = str(num)
print(str_num)  # Output: "10"

# Convert a float to a string
float_num = 2.718
str_float_num = str(float_num)
print(str_float_num)  # Output: "2.718"

# Convert a tuple to a list
my_tuple = (1, 2, 3)
my_list = list(my_tuple)
print(my_list)  # Output: [1, 2, 3]

# Convert a string to a list of characters
str_val = "hello"
char_list = list(str_val)
print(char_list)  # Output: ['h', 'e', 'l', 'l', 'o']

# Convert a list to a tuple
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print(my_tuple)  # Output: (1, 2, 3)

# Convert a string to a tuple of characters
str_val = "world"
char_tuple = tuple(str_val)
print(char_tuple)  # Output: ('w', 'o', 'r', 'l', 'd')

# Convert a list to a set (removes duplicates)
my_list = [1, 2, 2, 3]
my_set = set(my_list)
print(my_set)  # Output: {1, 2, 3}

# Convert a string to a set of characters
str_val = "banana"
char_set = set(str_val)
print(char_set)  # Output: {'a', 'b', 'n'}

```

```python
#Ans-6 #Here's a brief overview of conditional statements in Python, with comments and examples.

# Basic `if` Statement
age = 20
if age >= 18:
    print("You are an adult.")  # Output: You are an adult.

# `if` and `else` Statements
age = 16
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")  # Output: You are a minor.

# `if`, `elif`, and `else` Statements
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")  # Output: Grade: B
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: D")

# Nested `if` Statements
age = 22
if age >= 18:
    if age < 21:
        print("You can vote but not drink.")  # Output: You can vote but not drink.
    else:
        print("You can vote and drink.")
else:
    print("You are a minor.")

# Logical Operators in Conditions
temperature = 25
is_raining = False
if temperature > 20 and not is_raining:
    print("It's a nice day!")  # Output: It's a nice day!
```

```python
#Ans-7 # ` Here is a simplified overview of Python loops, with easy-to-understand comments and examples.



for` Loop: Loops over items in a sequence (like lists or ranges)
# Example: Looping through a list of numbers
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print(num)
# Output: 1 2 3 4 5

# Example: Looping through a range of numbers
for i in range(5):
    print(i)
# Output: 0 1 2 3 4

# `while` Loop: Repeats as long as a condition is true
# Example: Print numbers from 0 to 4
count = 0
while count < 5:
    print(count)
    count += 1
# Output: 0 1 2 3 4

# Nested Loops: Loop inside another loop, useful for tables or grids
# Example: Print coordinates in a 3x3 grid
for i in range(3):
    for j in range(3):
        print(f"({i}, {j})", end=" ")
    print()
# Output:
# (0, 0) (0, 1) (0, 2)
# (1, 0) (1, 1) (1, 2)
# (2, 0) (2, 1) (2, 2)

# `break`: Stops the loop immediately
# Example: Stop when you find the number 3
for num in [1, 2, 3, 4, 5]:
    if num == 3:
        break
    print(num)
# Output: 1 2

# `continue`: Skips the rest of the current loop iteration
# Example: Skip number 3
for num in [1, 2, 3, 4, 5]:
    if num == 3:
        continue
    print(num)
# Output: 1 2 4 5

# `else` with Loops: Runs after the loop finishes, if no `break` was hit
# Example: Check if 6 is in the list
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    if num == 6:
        print("Found 6")
        break
else:
    print("6 not found")  # Output: 6 not found
```
